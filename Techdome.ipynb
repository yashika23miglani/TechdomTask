```python
df['ShipDate'] = pd.to_datetime(df['ShipDate'], errors='coerce')
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'], errors='coerce')

# Sort data by ShipDate for time series plot
df.sort_values('ShipDate', inplace=True)

# Plotting ShipDate and DeliveryDate time series
plt.figure(figsize=(14, 7))
plt.plot(df['ShipDate'], df['ShippingCost'], label='Shipping Cost', marker='o')
plt.plot(df['ShipDate'], df['Revenue'], label='Revenue', marker='x')
plt.title('Shipping Cost and Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Cost / Revenue')
plt.legend()
plt.show()
```

```python
# Aggregating Revenue and Shipping Cost by ShipDate
daily_summary = df.groupby('ShipDate').agg({'Revenue': 'sum', 'ShippingCost': 'sum'})

# Plotting trends over time
plt.figure(figsize=(14, 7))
plt.plot(daily_summary.index, daily_summary['Revenue'], label='Total Revenue', marker='o', linestyle='-', color='blue')
plt.plot(daily_summary.index, daily_summary['ShippingCost'], label='Total Shipping Cost', marker='x', linestyle='-', color='red')
plt.title('Trends of Revenue and Shipping Cost Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Pie Chart for TripType

# Counting the instances of each TripType
trip_type_counts = df['TripType'].value_counts()

# Pie chart for TripType
plt.figure(figsize=(8, 8))
plt.pie(trip_type_counts, labels=trip_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Trip Types')
plt.show()
```

```python
#Bar Chart for Average Shipping Cost by OriginState
# Calculating average Shipping Cost by OriginState
average_shipping_by_state = df.groupby('OriginState')['ShippingCost'].mean()

# Bar chart for average Shipping Cost by OriginState
plt.figure(figsize=(12, 6))
average_shipping_by_state.sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Shipping Cost by Origin State')
plt.ylabel('Average Shipping Cost')
plt.xlabel('Origin State')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert date columns to datetime
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'])

# Calculate KPIs
# 1. Average Shipping Cost per Mile
df['CostPerMile'] = df['ShippingCost'] / df['TotalMiles']

# 2. Revenue per Loaded Mile
df['RevenuePerLoadedMile'] = df['Revenue'] / df['LoadedMiles']

# 3. Utilization Rate
df['UtilizationRate'] = df['LoadedMiles'] / df['TotalMiles'] * 100

# 4. On-time Delivery Rate
# Assuming 'DeliveryDate' is the actual delivery and 'ExpectedDeliveryDate' is when it was supposed to be delivered
# Adding a placeholder for 'ExpectedDeliveryDate' to demonstrate the calculation:
df['ExpectedDeliveryDate'] = df['DeliveryDate'] - pd.to_timedelta(1, unit='d')  # Placeholder logic
df['OnTimeDelivery'] = df['DeliveryDate'] <= df['ExpectedDeliveryDate']

# 5. Cost Variance
# Assuming 'ExpectedShippingCost' as a placeholder for demonstration:
df['ExpectedShippingCost'] = df['ShippingCost'] * 0.95  # Placeholder logic
df['CostVariance'] = df['ShippingCost'] - df['ExpectedShippingCost']

# Display calculated KPIs
kpis = df[['CostPerMile', 'RevenuePerLoadedMile', 'UtilizationRate', 'OnTimeDelivery', 'CostVariance']]
kpis.head()
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Widgets
start_date_picker = widgets.DatePicker(
    description='Start Date',
    value=min(df['ShipDate']),
    disabled=False
)

end_date_picker = widgets.DatePicker(
    description='End Date',
    value=max(df['ShipDate']),
    disabled=False
)

category_dropdown = widgets.Dropdown(
    options=['All'] + list(df['CategoryID'].unique()),
    value='All',
    description='Category:'
)

trip_type_dropdown = widgets.Dropdown(
    options=['All'] + list(df['TripType'].unique()),
    value='All',
    description='Trip Type:'
)

# Function to filter data
def filter_data(*args):
    filtered_df = df
    if category_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['CategoryID'] == category_dropdown.value]
    if trip_type_dropdown.value != 'All':
        filtered_df = filtered_df[filtered_df['TripType'] == trip_type_dropdown.value]
    if start_date_picker.value and end_date_picker.value:
        filtered_df = filtered_df[(filtered_df['ShipDate'] >= start_date_picker.value) & 
                                  (filtered_df['ShipDate'] <= end_date_picker.value)]
    display(filtered_df)

# Display widgets and bind the update function
display(start_date_picker, end_date_picker, category_dropdown, trip_type_dropdown)

# Observers
start_date_picker.observe(filter_data, names='value')
end_date_picker.observe(filter_data, names='value')
category_dropdown.observe(filter_data, names='value')
trip_type_dropdown.observe(filter_data, names='value')

# Initial display
filter_data()

```
